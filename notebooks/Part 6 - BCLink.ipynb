```python
import coconnect
import glob
```

```python
inputs =  coconnect.tools.load_csv(glob.glob('../data/part1/*'),nrows=1000)
inputs
```

```python
outputs = coconnect.io.BCLinkDataCollection({'dry_run':True},
                                            output_folder="./cache/",
                                            write_separate=True)
outputs
```

```python
rules = coconnect.tools.load_json("../data/rules.json")
```

```python
cdm = coconnect.cdm.CommonDataModel.from_rules(rules,inputs=inputs,outputs=outputs)
cdm.process()
```
