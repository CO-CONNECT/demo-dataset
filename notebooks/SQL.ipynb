```python
import coconnect
import glob
```

```python
inputs =  coconnect.tools.load_csv(glob.glob('./data/part2/*'))
inputs
```

```python
rules = coconnect.tools.load_json("data/rules.json")
rules
```

```python
outputs = coconnect.tools.create_sql_store(connection_string="postgresql://localhost:5432/ExampleCOVID19DataSet",
                                          drop_existing=True)
```

```python
cdm = coconnect.cdm.CommonDataModel.from_rules(rules,inputs=inputs,outputs=outputs)
cdm.get_tables()
```

```python
cdm.process()
```
